# Encoding of the Dataset

```python
def encode_gender(x):
  if x.lower()[0]=='F':
    return 0
  elif x.lower()[0]=='M':
    return 1
  else:
    return 2
```

```python
df['Choose your gender']=df['Choose your gender'].apply(encode_gender)
```

```python
df.select_dtypes('object').columns
```

```python
{column: list(df[column].unique()) for column in df.select_dtypes('object').columns}
```

```python
binary_features = [
  'Marital status','Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?','Did you seek any specialist for a treatment?'
]

ordinal_features = [
  'What is your CGPA?'
]

nominal_features = [
  'What is your course?','Your current year of Study'
]
```

```python
def binary_encode(df, columns, positive_values):
    df = df.copy()
    for column, positive_value in zip(columns, positive_values):
        df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)
    return df
def ordinal_encode(data, columns, orderings):
    data = data.copy()
    for column, ordering in zip(columns, orderings):
        df[column] = df[column].apply(lambda x: ordering.index(x))
    return df

def onehot_encode(data, columns, prefixes):
    data = data.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix)
        data = pd.concat([data, dummies], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
binary_positive_values = ['Yes' for feature in binary_features]
```

```python
ordinal_orderings = [
    ['3.00 - 3.49',
  '3.50 - 4.00',
  '3.50 - 4.00 ',
  '2.50 - 2.99',
  '2.00 - 2.49',
  '0 - 1.99']
]

nominal_prefixes = [
   
    'Wh',
    'Yo'
]
```

```python
df = binary_encode(
    df,
    columns=binary_features,
    positive_values=binary_positive_values
)
df=ordinal_encode(
    
   df,
   columns=ordinal_features,
   orderings=ordinal_orderings
)
df=onehot_encode(
    
   df,
   columns=nominal_features,
   prefixes=nominal_prefixes
)
```
